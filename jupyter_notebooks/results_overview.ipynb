## Joining and plotting results based on metric

### Some functions

```python
def separate_metric_result(metric_choice, calc_mean: bool = False, **dataframes, ):
    import re

    result = []
    for name, df in dataframes.items():
        if calc_mean:
            df = df.groupby('double', as_index=False).mean(numeric_only=True)
        if any(metric_choice in col for col in df.columns):
            regex_pattern = metric_choice
            if 'lgem' in name:
                # Firs: '_op'
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                filtered_metrics = filtered_metrics.filter(regex='op').copy()
                filtered_metrics.columns = [re.sub(r'_op$', '', col) for col in filtered_metrics.columns]
                joined_df = pd.concat([df.iloc[:, 0], filtered_metrics], axis=1)
                joined_df['Tool'] = f"{name} Optimized"
                result.append(joined_df)
                # Then: '_learn'
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                filtered_metrics = filtered_metrics.filter(regex='learn').copy()
                filtered_metrics.columns = [re.sub(r'_learn$', '', col) for col in filtered_metrics.columns]
                joined_df = pd.concat([df.iloc[:, [0]], filtered_metrics], axis=1)
                joined_df['Tool'] = f"{name} Learned"
                result.append(joined_df)

            else:
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                joined_df = pd.concat([df.iloc[:, 0], filtered_metrics], axis=1)
                joined_df['Tool'] = name
                result.append(joined_df)
    
    return pd.concat(result, ignore_index=True)
```

```python
def separate_val_result(metric_choice, **dataframes):
    import re

    result = []
    for name, df in dataframes.items():
        df = df.groupby('double', as_index=False).std(numeric_only=True)
        if any(metric_choice in col for col in df.columns):
            regex_pattern = metric_choice
            if 'lgem' in name:
                # Firs: '_op'
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                filtered_metrics = filtered_metrics.filter(regex='op').copy()
                filtered_metrics.columns = [re.sub(r'_op$', '', col) for col in filtered_metrics.columns]
                joined_df = pd.concat([df.iloc[:, 0], filtered_metrics], axis=1)
                joined_df['Tool'] = f"{name} Optimized"
                result.append(joined_df)
                # Then: '_learn'
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                filtered_metrics = filtered_metrics.filter(regex='learn').copy()
                filtered_metrics.columns = [re.sub(r'_learn$', '', col) for col in filtered_metrics.columns]
                joined_df = pd.concat([df.iloc[:, [0]], filtered_metrics], axis=1)
                joined_df['Tool'] = f"{name} Learned"
                result.append(joined_df)

            else:
                filtered_metrics = df.filter(regex=regex_pattern).copy()
                joined_df = pd.concat([df.iloc[:, 0], filtered_metrics], axis=1)
                joined_df['Tool'] = name
                result.append(joined_df)
    
    return pd.concat(result, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_metric(df, stripplot: bool = False, title: str = None, inner: str = None, cld_letters=None):

    df = df.iloc[:, 1:]
    # Assume df has: Tool, blue_sky, blue_ocean, etc.

    value_vars = [col for col in df.columns if col not in ['Tool']]  # all metric columns
    num_tools = len(df['Tool'].unique())

    df_melted = df.melt(id_vars='Tool', value_vars=value_vars,
                        var_name='Metric', value_name='Value')

    sns.set_theme(style="whitegrid", palette="pastel", color_codes=True)
    # Plot: grouped by Tool and Metric
    width_graph = np.ceil(num_tools*1.6)
    fig = plt.figure(figsize=(width_graph, 5))
    ax = plt.subplot(111)

    if stripplot == True:
        sns.violinplot(data=df_melted, x='Tool', y='Value', hue='Metric',
                    inner=None, dodge=True)
        sns.stripplot(data=df_melted, x='Tool', y='Value', hue='Metric',
                    dodge=True, jitter=True, marker='o', alpha=0.6, edgecolor='gray', linewidth=0.2)
    else:
        sns.violinplot(data=df_melted, x='Tool', y='Value', hue='Metric', dodge=True, split=False, width = 0.5, inner = inner)

    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(df_melted['Metric'].unique()))

    if title is not None:
        plt.title(title, pad=40)
    else:
        plt.title('Distribution of metrics by Tool', pad=40)

    if cld_letters is not None:
        # Get max value for each Tool to position the letter above the violin
        max_vals = df_melted.groupby('Tool')['Value'].max()
        for i, tool in enumerate(df['Tool'].unique()):
            y = max_vals[tool] * 1.05  # 5% above max
            ax.text(i, y, cld_letters.get(tool, ''), ha='center', va='bottom', fontsize=12, fontweight='bold')


    # plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_box(df, stripplot: bool = False, title: str = None, cld_letters=None):

    df = df.iloc[:, 1:]
    # Assume df has: Tool, blue_sky, blue_ocean, etc.

    value_vars = [col for col in df.columns if col not in ['Tool']]  # all metric columns
    num_tools = len(df['Tool'].unique())

    df_melted = df.melt(id_vars='Tool', value_vars=value_vars,
                        var_name='Metric', value_name='Value')

    sns.set_theme(style="whitegrid", palette="pastel", color_codes=True)
    # Plot: grouped by Tool and Metric
    width_graph = np.ceil(num_tools*1.6)
    fig = plt.figure(figsize=(width_graph, 5))
    ax = plt.subplot(111)

    if stripplot == True:
        sns.boxplot(data=df_melted, x='Tool', y='Value', hue='Metric',
                    inner=None, palette='pastel', dodge=True)
        sns.stripplot(data=df_melted, x='Tool', y='Value', hue='Metric',
                    dodge=True, jitter=True, marker='o', alpha=0.6, edgecolor='gray', linewidth=0.2)
    else:   
        sns.boxplot(data=df_melted, x='Tool', y='Value', hue='Metric', dodge=True, width=0.2)

    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(df_melted['Metric'].unique()))

    if title is not None:
        plt.title(title, pad=40)
    else:
        plt.title('Distribution of metrics by Tool', pad=40)

    if cld_letters is not None:
        # Get max value for each Tool to position the letter above the violin
        max_vals = df_melted.groupby('Tool')['Value'].max()
        for i, tool in enumerate(df['Tool'].unique()):
            y = max_vals[tool] * 1.05  # 5% above max
            ax.text(i, y, cld_letters.get(tool, ''), ha='center', va='bottom', fontsize=12, fontweight='bold')

    # plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_scatter(df, metric_name='example_metric', mean_value=False, complete_tools=False, choose_palette=None):

    if complete_tools:
        all_tools = set(df['Tool'].unique())
        tools_per_sample = df.groupby('double')['Tool'].apply(set)
        complete_samples = tools_per_sample[tools_per_sample == all_tools].index
        df = df[df['double'].isin(complete_samples)]

    # Alphabetically sort the sample names
    ordered_samples = sorted(df['double'].unique())
    df['double'] = pd.Categorical(df['double'], categories=ordered_samples, ordered=True)

    # For height stuff
    height_per_sample = 0.3
    num_samples = df['double'].nunique()
    fh = max(6, height_per_sample * num_samples)

    # Make sure color palette is the same for each Tool
    tools = sorted(df['Tool'].unique())
    if choose_palette is not None:
        tool_color_dict = choose_palette
    else:
        palette = sns.color_palette("Set1", n_colors=len(tools))
        tool_color_dict = dict(zip(tools, palette))


    # Melt the dataframe
    value_vars = [col for col in df.columns if col not in ['double', 'Tool']]
    df_melted = df.melt(id_vars=['double', 'Tool'], value_vars=value_vars,
                        var_name='Metric', value_name='Value')
    df_melted['double'] = pd.Categorical(df_melted['double'], categories=ordered_samples, ordered=True)

    # Plotting
    fig = plt.figure(figsize=(12, fh))
    ax = plt.subplot(111)
    
    # Lines for value in same Tool
    sns.lineplot(data=df_melted, x='Value', y='double', hue='Tool',
                 estimator=None, units='double', lw=1.5, legend=False, alpha=0.5, palette=tool_color_dict)

    if mean_value:
        # If you want to plot the mean value for each double
        df_means = df_melted.groupby(['double', 'Tool'], as_index=False)['Value'].mean()
        df_means['double'] = pd.Categorical(df_means['double'], categories=ordered_samples, ordered=True)
        sns.scatterplot(data=df_means, x='Value', y='double', hue='Tool', s=50, palette=tool_color_dict)

    else:
        # Dots for each Tool in each double
        sns.scatterplot(data=df_melted, x='Value', y='double', hue='Tool', s=50, palette=tool_color_dict)

    # Put legend outside
    box = ax.get_position()
    ax.set_position([box.x0, box.y0,
                    box.width, box.height * 0.9])

    # Get handles and labels of legend and order them as the palette given (or dict if not given)
    handles, labels = plt.gca().get_legend_handles_labels()
    order = list(tool_color_dict.keys())
    handles = [handles[labels.index(k)] for k in order if k in labels]
    labels = [k for k in order if k in labels]
    ax.legend(handles, labels,
              loc='upper right', bbox_to_anchor=(1, 1.05),
              fancybox=True, shadow=False, ncol=2)
    
    ax.set_ylabel("Perturbation condition")

    # Add vertical lines for visualisation
    plt.axvline(0, color='r', alpha=0.4, linestyle='--')

    # Customize the plot
    plt.title(f'{metric_name} by Perturbation Pair')
    # plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    plt.show()

    return df_melted

```

```python
def order_pert(double):
    perts = double.split('+')
    perts_sorted = sorted(perts)
    return '+'.join(perts_sorted)
```

```python
def summarize_model_performance(df, metric_cols, guide_column, hyperparams, decimals = 4):
    """
    Summarizes mean performance metrics for different hyperparameter settings.

    Parameters:
    - df: DataFrame containing results
    - metric_cols: List of metric column names, e.g., ['MSE', 'R2']
    - hyperparams: List of hyperparameter columns to group by, e.g., ['model', 'hidden_size']

    Returns:
    - summary_df: DataFrame with metrics as rows, hyperparameter combos as columns
    """
    # Group by the hyperparameters and compute mean metrics
    
    grouped = df.groupby([guide_column] + hyperparams)[metric_cols].T.mean()
    # Mean = mean by guide column
    # Var = depends on mean by guide column and overall mean for hyperparameter combos

    summary_mean = grouped.groupby(level=hyperparams).mean()
    summary_std = grouped.groupby(level=hyperparams).std(ddof=1)  # sample variance

    summary = summary_mean.copy()
    for metric in metric_cols:
        summary[metric] = summary_mean[metric].round(decimals).astype(str) + " Â± " + summary_std[metric].round(decimals).astype(str)
    summary = summary[metric_cols].T

    # summary_mean.index = pd.MultiIndex.from_product([summary_mean.index, ['mean']])
    # summary_std.index = pd.MultiIndex.from_product([summary_var.index, ['std']])
    # summary = pd.concat([summary_mean, summary_var])

    # Optional: flatten column MultiIndex into strings
    summary.columns = ['_'.join(map(str, col)) if isinstance(col, tuple) else str(col)
                       for col in summary.columns]

    return summary
```
