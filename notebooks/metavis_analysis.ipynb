````python
import scanpy as sc
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Leiden clustering on the original data (df_t)
# -----------------------
resolution = 1

adata = sc.AnnData(df_t)
sc.pp.neighbors(adata, n_neighbors=10, use_rep="X")
sc.tl.leiden(adata, resolution=resolution)
clusters = adata.obs["leiden"].astype(int).values
````

````python
def differential_analysis_subsystems_clusters_vs_rest(df, clusters):
    """
    For each cluster, perform a two-group comparison between:
    1) samples in the current cluster, and
    2) samples in all other clusters combined.

    Parameters
    ----------
    df : pd.DataFrame
        Rows are subsystems (string labels), columns are samples (numeric)
    clusters : array-like of int
        An array of cluster assignments for each column in df

    Returns
    -------
    pd.DataFrame
        A DataFrame with columns:
        ['subsystem', 'cluster', 't_value', 'p_value'],
        containing a row for each subsystem-cluster comparison.
    """
    import pandas as pd
    import numpy as np
    from scipy.stats import ttest_ind

    # Ensure df is numeric
    df = df.apply(pd.to_numeric, errors="coerce")

    unique_clusters = np.unique(clusters)
    results = []

    for subsystem in df.index:
        # Numeric row index for subsystem
        row_idx = df.index.get_loc(subsystem)

        for c in unique_clusters:
            # Indices for this cluster
            c_indices = np.where(clusters == c)[0]
            # Indices for all other clusters
            other_indices = np.where(clusters != c)[0]

            # Extract numeric values
            c_values = df.iloc[row_idx, c_indices].astype(float).dropna()
            others_values = df.iloc[row_idx, other_indices].astype(float).dropna()

            # Perform a two-sample t-test (assuming unequal variance)
            if len(c_values) > 1 and len(others_values) > 1:
                t_val, p_val = ttest_ind(c_values, others_values, equal_var=False)
            else:
                t_val, p_val = float("nan"), float("nan")

            results.append([subsystem, c, t_val, p_val])

    results_df = pd.DataFrame(
        results, columns=["subsystem", "cluster", "t_value", "p_value"]
    )
    results_df.sort_values("p_value", inplace=True)
    return results_df
````

````python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP

####################
# PCA
####################
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_t.values)

####################
# t-SNE
####################
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(df_t.values)

####################
# UMAP
####################
umap = UMAP(n_components=2, random_state=42)
umap_result = umap.fit_transform(df_t.values)
````

````python
def plot_clusters(
    embedding,
    clusters,
    unique_clusters=None,
    color_map=plt.cm.get_cmap("rainbow"),
    default_fig_size=(10, 8),
    label="PCA",
    ctrl_index=None,
):
    """Plot clustering results on a 2D embedding. Optionally highlight one sample (ctrl_index).

    Parameters
    ----------
    embedding : np.ndarray
        The 2D coordinates of the samples (n_samples, 2).
    clusters : array-like
        Cluster assignments for each sample.
    unique_clusters : array-like, optional
        Unique cluster labels. If None, inferred from 'clusters'.
    color_map : matplotlib.colors.Colormap, optional
        Colormap to use for clusters.
    default_fig_size : tuple, optional
        Figure size.
    label : str, optional
        Label for the embedding method (e.g., 'PCA', 't-SNE', 'UMAP').
    ctrl_index : int, optional
        The index of the sample to highlight. If specified, that point is shown with
        a different marker and labeled on the plot.

    Returns;
    -------
    cluster_color_assignments : list
        List of colors assigned to each cluster.
    """
    if unique_clusters is None:
        unique_clusters = np.unique(clusters)

    color_map = plt.cm.get_cmap("rainbow", len(unique_clusters))

    # Assign a unique color to each cluster
    cluster_color_assignments = [color_map(i) for i in unique_clusters]
    cluster_color_map = {
        uc: color for uc, color in zip(unique_clusters, cluster_color_assignments)
    }

    # Map each sample's cluster to its matching color
    sample_colors = [cluster_color_map[c] for c in clusters]

    plt.figure(figsize=default_fig_size)
    plt.scatter(embedding[:, 0], embedding[:, 1], c=sample_colors, marker="x", s=50)

    # Highlight the 'ctrl' sample if specified
    if ctrl_index is not None and 0 <= ctrl_index < embedding.shape[0]:
        plt.scatter(
            embedding[ctrl_index, 0],
            embedding[ctrl_index, 1],
            c="black",
            marker="o",
            s=30,
            label="ctrl",
        )
        plt.text(
            embedding[ctrl_index, 0],
            embedding[ctrl_index, 1] - 0.02,
            "ctrl",
            fontsize=9,
            ha="center",
            va="top",
            color="black",
        )

    plt.title(f"{label} of Reactions Activity (Leiden Clusters)")
    plt.xlabel(f"{label} Component 1")
    plt.ylabel(f"{label} Component 2")

    # Create custom legend entries
    handles = []
    for uc, color in zip(unique_clusters, cluster_color_assignments):
        handles.append(
            plt.Line2D(
                [],
                [],
                marker="x",
                color=color,
                linestyle="None",
                markersize=10,
                label=f"Cluster {uc}",
            )
        )

    if ctrl_index is not None and 0 <= ctrl_index < embedding.shape[0]:
        handles.append(
            plt.Line2D(
                [],
                [],
                marker="o",
                color="black",
                linestyle="None",
                markersize=10,
                label="ctrl",
            )
        )

    plt.legend(handles=handles, title="Clusters")
    plt.show()
````
